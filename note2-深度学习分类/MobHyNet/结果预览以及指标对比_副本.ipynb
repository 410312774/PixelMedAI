```python
from pixelmed_calc.medical_imaging.RadiologyComponents.components1 import HosmerLemeshow,draw_predict_score
from pixelmed_calc.medical_imaging.Ploting.plot_metric import plot_calibration_curves
model_names = pred_scores.keys()

pred_scores_train = [pred_scores[model]['train'][:,1] for model in model_names]
y_test_arrays = [label_dict[model]['train'] for model in model_names]

plot_calibration_curves(y_test_arrays,pred_scores_train,models=model_names,title='Train')
plt.savefig('results/calibration_curve_train.pdf',dpi=300)
```

```python


# 定义要进行hosmer_lemeshow_test的数据标签列表
label_groups = pred_scores.keys()
hosmer_test_results = []  


for group in label_groups:  
    test_labels = label_dict[group]['train']  
    test_scores = pred_scores[group]['train'][:, 1]  

    # 获取 HosmerLemeshow 结果，假设其本身是一个 DataFrame  
    hl_result_train = HosmerLemeshow(test_labels, test_scores, Q=80)  

    # 添加一个新的列标记 group  
    hl_result_train['Group'] = f"{group} train"  
    # 将结果追加到列表  
    hosmer_test_results.append(hl_result_train)  

    draw_predict_score(pred_scores[group]['train']  , pd.DataFrame(label_dict[group]['train'])  )
    plt.title(f'{group} sample predict score')
    plt.legend(labels=["label=0","label=1"],loc="lower right") 
    plt.savefig(f'results/model_{group}_sample_dis_train.svg', bbox_inches = 'tight')
    plt.show()
    plt.close()

# 将所有数据帧合并为一个  
results_df = pd.concat(hosmer_test_results, ignore_index=True)  

# 打印合并后的 DataFrame  
results_df
```

```python

def plot_multiple_PRs1(y_truths, y_scores, models=None, verbose=True, returnplot=False, title=None, lw=2):
    '''
    Compute and plot Precision-Recall (PR) curves for multiple estimators.

    Parameters:
    y_truths: list of arrays of ground truths, each array corresponding to an estimator
    y_scores: list of arrays of scores or labels, each array corresponding to an estimator
    models: list of model names, optional
    verbose: boolean, whether to print out messages
    returnplot: boolean, whether to return the plot object
    title: optional title string
    lw: line width

    Returns:
    precisions_list, recalls_list, thresholds_list: lists of arrays for each estimator
    '''
    import numpy as np
    import matplotlib.pyplot as plt
    from sklearn.metrics import precision_recall_curve, average_precision_score

    all_precisions = []
    all_recalls = []
    all_thresholds = []

    if models is None:
        models = [None] * len(y_truths)

    for y_truth, y_score, model in zip(y_truths, y_scores, models):
        y_truth = np.asarray(y_truth).astype(int)
        y_score = np.asarray(y_score)

        # 基本的类别存在性检查
        binc = np.bincount(y_truth)
        if len(binc) == 1 or (binc[0] == 0 or (len(binc) > 1 and binc[1] == 0)):
            print('[WORC Warning] One of the classes is missing; cannot determine PR. Skipping iteration.')
            continue

        # 情况1：y_score看起来是二值标签（只有两个唯一值）
        if y_score.ndim == 1 and np.unique(y_score).size == 2:
            # 将其当作分数的退化版本：依然可以计算PR，但阈值有限
            precisions, recalls, thresholds = precision_recall_curve(y_truth, y_score)
            ap_value = average_precision_score(y_truth, y_score)

            if verbose or returnplot:
                plt.plot(recalls, precisions, lw=lw, label=f'{model} (AP = {ap_value:0.3f})')

            all_precisions.append(precisions)
            all_recalls.append(recalls)
            all_thresholds.append(thresholds)
            continue

        # 情况2：连续分数，直接用sklearn计算
        precisions, recalls, thresholds = precision_recall_curve(y_truth, y_score)
        ap_value = average_precision_score(y_truth, y_score)

        all_precisions.append(precisions)
        all_recalls.append(recalls)
        all_thresholds.append(thresholds)

        if verbose or returnplot:
            plt.plot(recalls, precisions, lw=lw, label=f'{model} (AP = {ap_value:0.3f})')

    if verbose or returnplot:
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        if title is not None:
            plt.title(f'{title} PR')
        else:
            plt.title('Precision-Recall Curve')
        plt.grid(False)
        plt.legend(loc="lower left")

        if returnplot:
            return plt

    return all_precisions, all_recalls, all_thresholds
```

```python
from matplotlib.colors import Normalize  
from matplotlib.cm import ScalarMappable 
# 创建斯皮尔曼相关系数矩阵
results = {}  

# Compute p-values for each dataset type  
for dataset_type in dataset_types:  
    p_values = np.zeros((len(model_names), len(model_names)))  
    for i, (model_one, model_two) in enumerate(combinations(model_names, 2)):  
        ground_truth = np.array(label_dict[model_two][dataset_type])  
        predictions_one = pred_scores[model_one][dataset_type][:, 1]  
        predictions_two = pred_scores[model_two][dataset_type][:, 1]  

        delong = delong_roc_test(ground_truth, predictions_one, predictions_two)  
        p_value = delong[0][0]  

        # Fill symmetric matrix  
        idx_one = model_names.index(model_one)  
        idx_two = model_names.index(model_two)  
        p_values[idx_one][idx_two] = p_value  
        p_values[idx_two][idx_one] = p_value  

    # Convert to DataFrame for better visualization  
    results[dataset_type] = pd.DataFrame(p_values, index=model_names, columns=model_names)  

    # Create Bubble Heatmap  
    fig, ax = plt.subplots(figsize=(10, 8), dpi=1200)  
    cmap = plt.cm.viridis  
    norm = Normalize(vmin=0, vmax=1)  

    scatter_handles = []  
    for i in range(len(results[dataset_type].columns)):  
        for j in range(len(results[dataset_type].columns)):  
            if i > j:  # Lower triangle, show bubbles  
                color = cmap(norm(results[dataset_type].iloc[i, j]))  
                #scatter = ax.scatter(i, j, s=np.abs(results[dataset_type].iloc[i, j]) * 1000, color=color, alpha=0.75)  

                scatter = ax.scatter(i, j, s=(1.0 - results[dataset_type].iloc[i, j]) * 1000, color=color, alpha=0.75)  
                scatter_handles.append(scatter)  
            elif i < j:  # Upper triangle, show values  
                color = cmap(norm(results[dataset_type].iloc[i, j]))  
                ax.text(i, j, f'{results[dataset_type].iloc[i, j]:.2f}', ha='center', va='center', color=color, fontsize=10)  
            else:  # Diagonal  
                ax.scatter(i, j, s=1, color='white')  

    ax.set_xticks(range(len(results[dataset_type].columns)))  
    ax.set_xticklabels(model_names, rotation=45, ha='right', fontsize=10)  
    ax.set_yticks(range(len(results[dataset_type].columns)))  
    ax.set_yticklabels(model_names, fontsize=10)  
    
    # Add colorbar  
    sm = ScalarMappable(cmap=cmap, norm=norm)  
    sm.set_array([])  
    fig.colorbar(sm, ax=ax, label='P-value')  

    plt.title(f'Delong Test P-value Bubble Heatmap for {dataset_type.capitalize()} Dataset')  
    plt.tight_layout()  
    plt.savefig(f'Bubble_Heatmap_{dataset_type}.pdf', dpi=300, bbox_inches='tight')  
    plt.show()  
```
